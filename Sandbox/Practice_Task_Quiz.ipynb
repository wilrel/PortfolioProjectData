```python
# Data Analysis with Pandas and Matplotlib

This notebook reads a CSV file using `pandas`, processes the data, and plots a simple graph using `matplotlib.pyplot`.
```

```python
import pandas as pd

# Replace 'your_file.csv' with your actual CSV file path
df = pd.read_csv('Airline_Delay_Cause.csv')

# Display first 5 rows
df.head()
```

```python
import matplotlib.pyplot as plt

# Example: plot a line chart of two columns
plt.figure(figsize=(10, 6))
plt.plot(df['year'], df['carrier_name'], marker='o')
plt.title('Sample Plot')
plt.xlabel('Year')
plt.ylabel('Carrier Name')
plt.grid(True)
plt.show()
```
